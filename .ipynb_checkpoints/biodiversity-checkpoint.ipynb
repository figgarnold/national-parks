```python
#import first data source and view high level
species_info = pd.read_csv ("species_info.csv")
print(species_info.head())
```

```python
#high level visualisation of the data 
sns.countplot(species_info, x="category", hue="conservation_status")
```

```python
#remove the null items spotted before and check
species_info.fillna('not_target', inplace=True)
species_info.groupby("conservation_status").size()
```

```python
species_info.groupby("category").size()
```

```python
#create graph by category excluding the not target (previously null) values
plt.figure(figsize=(6, 8))

ocplot = sns.countplot(species_info[species_info.conservation_status != "not_target"], x="conservation_status", hue="category")
```

```python
#create table of counts 
species_info['protected'] = species_info.conservation_status != 'not_target'

category_counts = species_info.groupby(['category', 'protected'])\
                        .scientific_name.nunique()\
                        .reset_index()\
                        .pivot(columns='protected',
                                      index='category',
                                      values='scientific_name')\
                        .reset_index()
category_counts.columns = ['category', 'not_target', 'protected']

category_counts
```

```python
#get percentage per category
category_counts['percent_protected'] = category_counts.protected / \
                                      (category_counts.protected + category_counts.not_target) * 100

category_counts
```

```python
#statistical significance Test 1 Mammals Vs Birds

from scipy.stats import chi2_contingency

contingency1 = [[30, 146],
              [75, 413]]
chi2_contingency(contingency1)
```

```python
#p value is .69 between Mammals and Birds which means there is no statistical significance
```
